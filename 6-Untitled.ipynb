```python
text = open('6-news1925462.txt').read()
text
```

```python
sents = [line.strip() for line in text.splitlines()]
sents
```

# Tokenization

```python
from nltk import wordpunct_tokenize
list_of_words = [wordpunct_tokenize(sent) for sent in sents]
list_of_words
```

```python
words = [word for list_of_word in list_of_words for word in list_of_word]
words
```

```python
words = [word for word in words if word.isalpha() or word.isdigit() or word.isalnum()]
words
```

```python
words = [word.lower() for word in words]
words
```

```python
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
words = [word for word in words if word not in stopwords]
words
```

```python
from wordcloud import WordCloud
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

wordcloud = WordCloud(words)
plt.imshow(wordcloud)
plt.axis('off')
```
