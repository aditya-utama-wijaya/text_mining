## Looping with Conditions

```python
for xyzzy in sent1:
    if xyzzy.endswith('l'):
        print(xyzzy)
```

```python
for token in sent1:
    if token.islower():
        print(token, 'is a lowercase word')
    elif token.istitle():
        print(token, 'is a titlecase word')
    else:
        print(token, 'is punctuation')
```
