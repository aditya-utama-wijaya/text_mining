```python
import pandas as pd
theses = pd.read_csv('7-theses.csv')
theses
```

```python
def split_two_advisors(df):
    s = df["advisor"].str.split('\n', expand = True).stack()
    i = s.index.get_level_values(0)
    df2 = df.loc[i].copy()
    df2["advisor"] = s.values
    return df2
theses = split_two_advisors(theses)
theses
```

```python
theses = theses.sort(['department', 'advisor'])
theses
```

```python
advisors = [advisor for advisor in theses['advisor']]
advisors
```

```python
from nltk import wordpunct_tokenize
words_per_advisor = [wordpunct_tokenize(advisor) for advisor in advisors]
words_per_advisor
```

```python
for i, word in enumerate(words_per_advisor):
    words_per_advisor[i] = [w for w in words_per_advisor[i] if w.isalpha() or w.isdigit() or w.isalnum()]
words_per_advisor
```

```python
for i, word in enumerate(words_per_advisor):
    words_per_advisor[i] = [w.lower() for w in words_per_advisor[i]]
words_per_advisor
```

```python
theses.insert(2, 'advisor_tokenized', words_per_advisor)
theses
```

```python
theses.to_csv('7-theses2.csv', encoding = 'utf-8', index = False)
```

```python
theses = pd.read_csv('7-theses3.csv')
theses
```

```python
abstracts = [abstract for abstract in theses['abstract']]
abstracts[:5]
```

```python
words_per_abstract = [wordpunct_tokenize(abstract) for abstract in abstracts]
words_per_abstract[:5]
```

```python
for i, word in enumerate(words_per_abstract):
    words_per_abstract[i] = [w for w in words_per_abstract[i] if w.isalpha() or w.isdigit() or w.isalnum()]
words_per_abstract[:5]
```

```python
for i, word in enumerate(words_per_abstract):
    words_per_abstract[i] = [w.lower() for w in words_per_abstract[i]]
words_per_abstract[:5]
```

```python
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
for i, word in enumerate(words_per_abstract):
    words_per_abstract[i] = [w for w in words_per_abstract[i] if w not in stopwords]
words_per_abstract[:5]
```

```python
theses['abstract_tokenized'] = words_per_abstract
theses.sort(['advisor_tokenized'])
```

```python
theses.to_csv('7-theses_final.csv', encoding = 'utf-8', index = False)
```
