## Stopwords

Download nltk stopwords (Corpora -> stopwords)

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
```

```python
print(stopwords)
```

Filter stopwords

```python
words = [word for word in word_tokenize(sents[1]) if word not in stopwords]
print(words)
```

## Wordcloud

```python
cut_text = '\n'.join(' '.join(word_tokenize(sent)) for sent in sents)
cut_text
```

```python
from wordcloud import WordCloud
%matplotlib inline
import matplotlib.pyplot as plt

wordcloud = WordCloud(stopwords=stopwords).generate(cut_text)
plt.imshow(wordcloud)
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=40, stopwords=stopwords).generate(cut_text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
